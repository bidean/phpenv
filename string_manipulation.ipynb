```python
name = input("Enter your first name: ")
print(len(name))
surname = input("Enter your surname: ")
print(len(surname))
fullname = name+" "+surname
print(fullname)
print("The length of your fullname is: ", len(fullname), " characters")
```

```python
poem_line = "Mirror Mirror on the wall "
start = int(input("Enter a start point: "))
end = int(input("Enter an end point: "))

print(poem_line[start:end])
```

```python
word = input("Enter a word in upper case: ")

while word.islower:
  print("Try again")
  word = input("Enter a word in upper case: ")
print(word, "is now in upper case")

print(word.swapcase())

```
