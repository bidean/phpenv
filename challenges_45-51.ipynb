```python
compnum = 50
attempts = 0
guess = int(input("Enter a number: "))
attempts = attempts + 1
while guess != compnum:
  if guess < compnum:
    print("guess to low")
    attempts = attempts + 1
    guess = int(input("Enter a number: "))
  elif guess > compnum:
    print("guess to high")
    attempts = attempts + 1
    guess = int(input("Enter a number: "))
print("You guessed correctly ! You had ", attempts, "attempts")


```
