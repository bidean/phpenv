```python
#from IPython.core.inputtransformer2 import Compile
import random

#computer random colour
comp_colour = random.choice(["red", "green", "blue", "yellow", "purple"])

#user first guess
user_choice = input(" Guess the colour: red, green, blue, yellow, purple")

#set the global guess to true. This will determine if we need to ask the user
#again for another guess
guess = True

#if the user and computer matches:
if comp_colour == user_choice:
  print("Well done, you chose correctly!")
  guess = True

#only run this while the computer choice and user choice are different
#if they are different, give the user a clue... Yellow Submarine
while comp_colour != user_choice:
    print("Incorrect")
    guess = False

    #Give the user a clue
    if comp_colour == "green":
      user_choice = input("Clue: Froggy?")
    elif comp_colour == "red":
      user_choice = input("Clue: Barchetta? ")

    elif comp_colour == "yellow":
      user_choice = input("Clue .... Submarine?")

    elif comp_colour == "blue":
      user_choice = input("Clue: one for the money")

    elif comp_colour == "purple":
      user_choice = input("Clue: Prince?")


    #if their second guess is correct, say well done and change guess back to True
    #The clues will not run if the user guesses correctly the first time
    if user_choice == comp_colour:
      print("Well done. You guessed correctly!")
      guess = True
```

```python
# Example dictionary
my_dict = {
    "name": "Alice",
    "age": 30,
    "city": "New York",
    "occupation": "Engineer"
}
print(f"Original dictionary: {my_dict}")

# --- Using the del keyword ---
# Deleting an item by its key
del my_dict["age"]
print(f"Dictionary after using del on 'age': {my_dict}")

# --- Using the pop() method ---
# Removing an item and getting its value back
removed_city = my_dict.pop("city")
print(f"Dictionary after using pop('city'): {my_dict}")
print(f"Removed value (city): {removed_city}")

# Removing an item with a default value if the key doesn't exist
# This prevents a KeyError if 'occupation' is already gone
removed_occupation = my_dict.pop("occupation", "Not Found")
print(f"Dictionary after using pop('occupation') with default: {my_dict}")
print(f"Removed value (occupation): {removed_occupation}")

# Trying to pop a non-existent key without a default will raise an error
try:
  my_dict.pop("country")
except KeyError as e:
  print(f"Error: {e} - 'country' is not in the dictionary.")

# The clear() method removes all items from the dictionary
my_dict.clear()
print(f"Dictionary after using clear(): {my_dict}")
```

```python
digits = []

for i in range(0,4):
  digit = input(f"Enter a 3 digit number ({i+1}/4): ")
  digits.append(digit)

print("Current digits in the list:")
for n in digits:
  print(n)

number = input("Enter a 3 digit number: ")

while number not in digits:
  print("Incorrect!")
  number = input("Enter a 3 digit number: ")

print(f"The number '{number}' is in the list of digits at index {digits.index(number)}.")
```

```python
names = []
for i in range(0,3):
  # Correctly format the prompt using an f-string and avoid names.index(i)
  name = input(f"Enter the name of person {i+1} to invite: ")
  names.append(name)
print(names)

new_name = input("Do you want to invite another person?")

while new_name == "yes" or new_name == "Yes":
  name = input(f"Enter another person to invite: ")
  names.append(name)
  new_name = input("Do you want to invite another person?")
print("These are the final names that you have invited: ", names)

remove = input("Type in one of the names in the list")

# Check if the name to remove is actually in the list before trying to find its index
if remove in names:
  print(remove, "is in position:", names.index(remove), "on the list.")
else:
  print(f"'{remove}' is not in the list.")

keep = input("Do you still want to invite this person to the party?")
if keep == "no" or keep == "No":
  # Only attempt to remove if the name is still in the list
  if remove in names:
    names.remove(remove)
    print(remove, " was removed from the list")
  else:
    print(f"'{remove}' was already removed or not found.")

print("This is the final list:", names)
```
