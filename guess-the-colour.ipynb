```python
#from IPython.core.inputtransformer2 import Compile
import random

#computer random colour
comp_colour = random.choice(["red", "green", "blue", "yellow", "purple"])

#user first guess
user_choice = input(" Guess the colour: red, green, blue, yellow, purple")

#set the global guess to true. This will determine if we need to ask the user
#again for another guess
guess = True

#if the user and computer matches:
if comp_colour == user_choice:
  print("Well done, you chose correctly!")
  guess = True

#only run this while the computer choice and user choice are different
#if they are different, give the user a clue... Yellow Submarine
while comp_colour != user_choice:
    print("Incorrect")
    guess = False

    #Give the user a clue
    if comp_colour == "green":
      user_choice = input("Clue: Froggy?")
    elif comp_colour == "red":
      user_choice = input("Clue: Barchetta? ")

    elif comp_colour == "yellow":
      user_choice = input("Clue .... Submarine?")

    elif comp_colour == "blue":
      user_choice = input("Clue: one for the money")

    elif comp_colour == "purple":
      user_choice = input("Clue: Prince?")


    #if their second guess is correct, say well don and change guess back to True
    #The clues will not run if the user guesses correctly the first time
    if user_choice == comp_colour:
      print("Well done. You guessed correctly!")
      guess = True










```
